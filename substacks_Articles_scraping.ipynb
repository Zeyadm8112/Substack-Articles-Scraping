```python
import re
import time

def get_valid_csv_name(name):
    # Replace any non-alphanumeric characters (except for hyphens and underscores) with underscores
    name = re.sub(r'[^\w-]', '_', name)
    
    # Add a timestamp to the name to make it unique
    timestamp = time.strftime("%Y%m%d-%H%M%S")
    name = f"{name}_{timestamp}"
    
    # Add the ".csv" extension
    name = f"{name}.csv"
    
    return name
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

title_list=[]
link_list=[]

# Set up the Selenium web driver
driver = webdriver.Chrome()
df = pd.read_csv('F:\\upwork\\crypto\\2nd_milestone\\web3_publications.csv')

# Extract the values from the first and third columns
titles_list = df.iloc[:, 0].tolist()  # Column indices are 0-based
links_list = df.iloc[:, 2].tolist()

for i in range(183,len(titles_list)):
    print(i)
    print(titles_list[i])
    url = links_list[i]
    driver.get(url)
    try: 
        skip_button=driver.find_element(By.XPATH,'//*[@id="main"]/div[2]/div[3]/div/div/div/a/button')
        skip_button.click()
    except NoSuchElementException:
        print("Skip button not found, skipping...")
        
    try:            
        see_all_button=driver.find_element(By.XPATH,'//*[@id="home-body"]/div[1]/div/div[1]/div[2]/div[2]/button')
        see_all_button.click()
    except NoSuchElementException:
        print("See all button not found, skipping...")

    # Scroll down to load all the articles
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:    
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            print("Reached end of page")
            break
        last_height = new_height

        # Stop scrolling after reaching a certain point to avoid infinite loop
        if new_height > 100000:
            print("Reached maximum scroll height")
            break

    # Extract the HTML content of the page
    html_content = driver.page_source
  #  print(html_content)

    # Create a BeautifulSoup object from the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
   # print(soup)

    # Find the elements that contain the articles
    article_elements = soup.find_all('div', {'class': 'frontend-main-home-PostPreview-PostPreviewListing-module__container--zH3ui'})
   # print(article_elements)

    # Loop through the article elements and extract the title and content
    for article in article_elements:
        title = article.find("a", class_="pencraft frontend-pencraft-Box-module__reset--VfQY8 frontend-pencraft-Text-module__size-18--zqngu frontend-pencraft-Text-module__line-height-28--NgO1Q frontend-pencraft-Text-module__weight-semibold--LJBj3 frontend-pencraft-Text-module__font-pub-headings--lbOZ2 frontend-pencraft-Text-module__clamp--a1dYM frontend-pencraft-Text-module__clamp-3--R6uR3 frontend-pencraft-Text-module__reset--dW0zZ").text.strip()
        title_list.append(title)
        link = article.find("a", class_="pencraft frontend-pencraft-Box-module__reset--VfQY8 frontend-pencraft-Text-module__size-18--zqngu frontend-pencraft-Text-module__line-height-28--NgO1Q frontend-pencraft-Text-module__weight-semibold--LJBj3 frontend-pencraft-Text-module__font-pub-headings--lbOZ2 frontend-pencraft-Text-module__clamp--a1dYM frontend-pencraft-Text-module__clamp-3--R6uR3 frontend-pencraft-Text-module__reset--dW0zZ")["href"]
        link_list.append(link)
    print(len(title_list))
    #print(title_list)
   # print(link_list)
    
    adjusted_name = get_valid_csv_name(titles_list[i])
    with open(f'F:\\upwork\\crypto\\3rd_milestone\\web3_publications\\{adjusted_name}.csv', 'w', newline='',encoding='utf-8') as csvfile: 
        writer = csv.writer(csvfile)
    # Write the column headers
        writer.writerow(['Title', 'Link'])
    # Write the data rows
        for title, link in zip(title_list, link_list):
            
            writer.writerow([title, link])
# Close the web driver
driver.quit()
```
